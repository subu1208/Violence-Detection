```python
input_video_folder='E:\JANANI_PROJECT\data'
for file in os.listdir(input_video_folder)[600:700]:
  input_video_path=input_video_folder+'/'+file
  cap=cv2.VideoCapture(input_video_path)
  ret, first_frame =cap.read()
  prev_gray = cv2.cvtColor(first_frame, cv2.COLOR_BGR2GRAY)
  frame_width = int(cap.get(3))
  frame_height = int(cap.get(4))
  size = (frame_width, frame_height)
  out='change-'+file
  result = cv2.VideoWriter(out, cv2.VideoWriter_fourcc(*'MJPG'), 10, size)
  mask = np.zeros_like(first_frame)
  mask[..., 1] = 255
  while True:
    ret, frame = cap.read()
    if ret==True:
      gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
      flow = cv2.calcOpticalFlowFarneback(prev_gray, gray,None,0.5, 3, 15, 3, 5, 1.2, 0)
      magnitude, angle = cv2.cartToPolar(flow[..., 0], flow[..., 1])
      mask[..., 0] = angle * 180 / np.pi / 2
      mask[..., 2] = cv2.normalize(magnitude, None, 0, 255, cv2.NORM_MINMAX)
      rgb = cv2.cvtColor(mask, cv2.COLOR_HSV2BGR)
      result.write(rgb)
      prev_gray = gray
    else:
        break
  cap.release()
  result.release()
```

3.Fusion of pose estimation and change detection

```python
import cv2
import numpy as np
input_video_folder='E:\JANANI_PROJECT\data'
count=0
for file in os.listdir(input_video_folder)[600:700]:
  print(count)
  count+=1
  pose_video_path='E:\JANANI_PROJECT\pose\pose-'+file
  change_video_path='E:\JANANI_PROJECT\change-detection\change-'+file
  cap1=cv2.VideoCapture(pose_video_path)
  cap2=cv2.VideoCapture(change_video_path)
  frame_width = int(cap1.get(3))
  frame_height = int(cap2.get(4))
  size = (frame_width, frame_height)
  out='fusion-'+file
  result = cv2.VideoWriter(out, cv2.VideoWriter_fourcc(*'MJPG'), 10, size)
  while True:
    ret1,frame1=cap1.read()
    ret2,frame2=cap2.read()
    if ret1 and ret2:
      img = cv2.addWeighted(frame1, 0.4, frame2, 0.6, 0)
      result.write(img)
    else:
      break
  cap1.release()
  cap2.release()
  result.release()
```
